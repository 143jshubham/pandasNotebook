```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
match=pd.read_csv('matches (1).csv')
delivery=pd.read_csv('deliveries.csv')

ipl=delivery.merge(match,left_on='match_id',right_on='id')
    
```

```python
def performance(batsman):
    mask=ipl[ipl['batsman']==batsman]
    perform=mask.groupby('season')['batsman_runs'].sum()
    plt.plot(perform)

```

```python
performance('V Kohli')
```

```python
def runtype(batsman):
    mask=ipl[ipl['batsman']==batsman]
    perform=mask.groupby('batsman_runs')['batsman_runs'].count()
    plt.pie(perform,autopct="%0.1f%%")
    plt.show()
    
```

```python
runtype('MS Dhoni')
```
